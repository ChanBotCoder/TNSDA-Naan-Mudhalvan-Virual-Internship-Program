```python
while True:
  number=input("Enter the Roman number:")
  Roman_dict={"I":1,"V":5,"X":10,"L":50,"C":100,"D":500,"M":1000} 
  value=0
  for i in range(len(number)):
    if i>0 and Roman_dict[number[i]]>Roman_dict[number[i-1]]: 
      value+=Roman_dict[number[i]] - 2 * Roman_dict[number[i-1]]
    else:
      value+=Roman_dict[number[i]]
  print("Decimal value", value)
```
